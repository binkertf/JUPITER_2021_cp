# Running Jupyter Notebook/Lab on a remote server

On your remote machine, launch a Jupyter Notebook server:

```python
jupyter notebook --no-browser --port=<PORT>
```

Replace ```<PORT>``` with your selected port number.

or alternatively launch a Jupyter Lab server:

```python
jupyter lab --no-browser --port=<PORT>
```

I will now proceed with a Jupyter Notebook server. For a Jupyter Lab server it works exactly the same and it only depends on your preference which application you want to use. 

For example, if you want to use a Jupyter Notebook on port number 8893, you would run the following:

```python
jupyter notebook --no-browser --port=8893
```

Or run the following command to launch with default port:

```python
jupyter notebook --no-browser
```

On your local machine, e.g. your laptop, you can access the notebook from your remote machine over SSH by setting up a SSH tunnel. Run the following command from your local machine:

```python
ssh -N -f -L 8080:localhost:<PORT> <REMOTE_USER>@<REMOTE_HOST>
```

Replace ```<PORT>``` with the port number you selected in the above step \
Replace ```<REMOTE_USER>``` with the remote server username \
Replace ```<REMOTE_HOST>``` with your remote server address

Open a browser (e.g. firefox, chrome, ...) from your local machine and navigate to http://localhost:8893/, the Jupyter Notebook web interface. Replace 8893 if you used a different port number in step 1. Navigate to http://localhost:8893/lab if you want to access a Juypter Lab server instead.

````python
N = 100 #set the output number
````

````python
jup.output(N).plotmidslice(field='dustvelocity',polar=False,vel=2,z=0) #plot a slice parallel to the midplane
jup.output(N).plotvertslice(field='gasdensity',polar=False,azi=0.0,vel=0) #plot a vertical slice (perpendicular to the midplane)
jup.output(N).plotvertavg(field='gasdensity',polar=False) #plot the azimuthal average of all the vertical slices
jup.output(N).plotgapprofile(field='gasdensity',ylog=False,avg=True,azi=0.0) #plot the radial surface density profile
````
